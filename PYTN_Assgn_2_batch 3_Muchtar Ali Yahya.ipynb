Confidence Interval

```python
confidence_level = 0.95

# Tentukan jumlah sampel
n = len(data.LOT)

# Tentukan derajat kebebasan
df = n - 1

# Hitung nilai t dengan derajat kebebasan yang sesuai
t = stats.t.ppf(confidence_level, df)
print("t-value:", t)

```

```python
# Hitung Confidence Interval
ci = mean + (t * std / 1/(n^2))
print("Confidence Inerval: ", ci)
```
